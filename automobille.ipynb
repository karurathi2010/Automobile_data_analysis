```python
# Import dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# assigning file path.
path='./Data/automobile_data.csv'
```

```python
# Read csv file
auto_df=pd.read_csv(path)
```

```python
# Deleting unwanted column
del auto_df['Unnamed: 0']
```

```python
# Preview dataframe
auto_df.head()
```

```python
# Check Dataframe info.
auto_df.info()
```

```python
# Getting summary of the dataframe
auto_df.describe(include='all')
```

```python
# Replace null values with mean values
mean_value=auto_df['normalized-losses'].mean()
auto_df['normalized-losses'].fillna(mean_value,inplace=True)
```

```python
# Replace null values with mode.
doors=auto_df['num-of-doors'].mode()[0]
auto_df['num-of-doors'].fillna(doors,inplace=True)
```

```python
# Replace null values with mean values
price=auto_df['price'].mean()
auto_df['price'].fillna(price,inplace=True)
```

```python
# Droping null values
auto_df.dropna(inplace=True)
```

```python
auto_df.info()
```

```python
# binning the 'horsepower' column to make it categorical.
bins=np.linspace(min(auto_df['horsepower']),max(auto_df['horsepower']),4)
labels=['low','medium','high']
auto_df['binned_horsepower']=pd.cut(auto_df['horsepower'],bins=bins,labels=labels,include_lowest=True)
```

```python
auto_df['binned_horsepower'].value_counts()
```

```python
# pattern recognition
sns.regplot(x='engine-size',y='price',data=auto_df)
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Correlation Between Engine size and Price')
plt.show()
```

```python
# Checking the correlation coefficent
auto_df[['engine-size','price']].corr()
```

```python
# pattern recognition
sns.regplot(x='peak-rpm',y='price',data=auto_df)
plt.xlabel('Peak RPM')
plt.ylabel('Price')
plt.title('Correlation Between Peak RPM and Price')
plt.show()
```

```python
auto_df[['peak-rpm','price']].corr()
```

```python
# Checking the correlation between the numerical values
corel=auto_df.corr(numeric_only=True)
tick=corel.columns
```

```python
# Plotting correlation matrix
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
cax=ax.matshow(corel,cmap='coolwarm')
fig.colorbar(cax)
ticks=np.arange(0,len(tick),1)
plt.xticks(rotation=90)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(corel.columns)
ax.set_yticklabels(corel.columns)
plt.show()
```

```python
# Checking the relationship between body style and price
sns.boxplot(x='body-style',y='price',data=auto_df)
plt.xlabel('Body Style')
plt.ylabel('Price')
plt.title('Body Style Vs Price Plot')
plt.show()
```

From the above plot,it is clear that the body type doesn't have much influence in the price.

```python
# Checking the relationship engine location style and price
sns.boxplot(x='engine-location',y='price',data=auto_df)
plt.xlabel('Engine-Location')
plt.ylabel('Price')
plt.title('Engine-Location Vs Price Plot')
plt.show()
```

The above plot shows that the engine location is a good indicator of price.Rear engine vehicles are costly compared to the front engine vehicles.So Engine-Location is a good feature for predicting the price.

```python
new_df=auto_df[['drive-wheels','price']].groupby('drive-wheels')
new_df
```

```python
from scipy.stats import pearsonr
pearsonr(auto_df['wheel-base'],auto_df['price'])
```

```python
#ANOVA test
from scipy import stats as st
st.f_oneway(new_df.get_group('fwd')['price'],new_df.get_group('rwd')['price'],new_df.get_group('4wd')['price']) 
```

The above f_onway test shows that the wheel-base feature is good predictor of price.
